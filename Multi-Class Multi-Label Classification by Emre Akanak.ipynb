```python
%%time 

#uds: Unprocessd Data Shape
uds = data.shape[0]


# 1: 'Unnamed: 0' Column
# 'Unnamed: 0' Column is deleted 
del data['Unnamed: 0']

#-----------------------------------------------------------

# 2: 'Cust_ID' Column
# 'Cust_ID' represents customer ID and might be used as indexing and contains no NaN data.
# It stays as it is 

#-----------------------------------------------------------

# 3: 'Status_Dt' Column
# 'Status_Dt' was transformed to datetime. It is not really necessary but seems better to do it anyway 
data['Status_Dt'] = pd.to_datetime(data['Status_Dt'], format = '%Y-%m-%d')
data['Status_Dt'] = data['Status_Dt'].dt.normalize()

data['month'] = data['Status_Dt'].dt.month_name()

#-----------------------------------------------------------

# 4: 'Gross_HHLD_Income' Column:
# 'Gross_HHLD_Income' Value was fillet by mean value according to categories.
# Multiiple criterias was used in order to reduce potential biases. 'Age', 'Cust_Province_Name' and 'Cust_Sex' are used.
data['Gross_HHLD_Income'] = data.groupby(['Age', 'Cust_Province_Name', 'Cust_Sex'])['Gross_HHLD_Income'].transform(lambda x: x.fillna(x.mean()))

# Income Group Calculation

Q1 = data['Gross_HHLD_Income'].quantile(0.25)
Q3 = data['Gross_HHLD_Income'].quantile(0.75)
IQR = Q3 - Q1

# Income Categorization Methodology
# Income               Category 
# 0 - Q1              Low Income Group
# Q1 - IQR            Middle Income Group
# IQR - Q3            High Income Group
# Q3 - Max            Affulent Income Group 


#cid: Categorized Income Data
data['cid'] = pd.cut(x=data['Gross_HHLD_Income'], bins = [0, Q1, IQR, Q3, data['Gross_HHLD_Income'].max()], labels=['Low Income', 'Middle Income', 'High Income', 'Affulent'])






#-----------------------------------------------------------

# 5: 'Cust_Province_Name' Colum:
# I checked the provincies and clients concantrated in some speccific areas.
# ZARAGOZA and SANTA CRUZ DE TENERIFE seems there are outliers about mean income but in general there is no speccific weight
# So I decided to categorize and numerize the data as ordinal concept by clinet density:
# Ordinal Enumerating Details: 
# Province             Percentage of Clients        Encoding Number
# MADRID                    33.67%                         1
# BARCELONA                 9.28%                          2
# VALENCIA                  5.18%                          3
# SEVILLA                   4.38%                          4
# CORUA, A                 2.88%                          5
# MURCIA                    2.86%                          6
# ZARAGOZA                  2.55%                          7
# MALAGA                    2.47%                          8
# ALICANTE                  2.41%                          9
# CADIZ                     2.27%                          10 
# Others                    Less than 1%                   0 

# epd: Encoded Province Data
provincescalemapper = {'MADRID': 1,'BARCELONA': 2, 'VALENCIA': 3, 'SEVILLA': 4, 'CORUA, A': 5, 'MURCIA': 6, 'ZARAGOZA': 7, 'MALAGA': 8, 'ALICANTE': 9,'CADIZ':10,'HUESCA':0,'CIUDAD REAL':0,'SALAMANCA':0,'TOLEDO':0,'RIOJA, LA':0,'PONTEVEDRA':0,'HUELVA':0,'LERIDA':0,'VALLADOLID':0,'BURGOS':0,'JAEN':0,'CASTELLON':0,'CORDOBA':0,'SANTA CRUZ DE TENERIFE':0,'CACERES':0,'LUGO':0,'TARRAGONA':0,'PALENCIA':0,'ZAMORA':0,'ALBACETE':0,'BADAJOZ':0,'AVILA':0,'GIRONA':0,'ALMERIA':0,'GRANADA':0,'OURENSE':0,'BALEARS, ILLES':0,'SEGOVIA':0,'CANTABRIA':0,'ASTURIAS':0,'CUENCA':0,'LEON':0,'PALMAS, LAS':0,'CEUTA':0,'GUADALAJARA':0,'TERUEL':0,'SORIA':0,'MELILLA':0,'GIPUZKOA':0}
data['epd'] = data['Cust_Province_Name']
data['epd'] = data['epd'].replace(provincescalemapper)


#-----------------------------------------------------------


# 6: 'Join_Channel'Column:
# It openly shows most of the clients came from 4 (four) channels. KHE, KAT, KFC and KFA totlly  85% of the total clients.
# The encoding methodology is as follow:
# Join Channel Name            Total Client percentage                 Encode
#       KHE                          32.18%                              1
#       KAT                          26.93%                              2
#       KFC                          23.01%                              3
#       KFA                          3.16%                               4
#       Others                                                           0

# jce: Joining Channel Encoder

data['channel'] = data['Join_Channel']

channelscalemapper = {'KHE':1, 'KAT':2, 'KFC':3, 'KFA':4, 'KHD':0, 'KHC':0, 'KAZ':0, 'KHK':0, 'KAE':0, 'KFU':0, 'KCC':0, 'KAA':0, 'KAB':0, 'KFG':0, 'KAG':0, 'KFD':0, 'KFN':0, 'KFP':0,'RED':0,'KHF':0,'KHL':0,'KBG':0,'KDE':0,'KDR':0,'KBO':0,'KCL':0,'KBQ':0,'KCB':0,'KCG':0,'KES':0,'KCD':0,'KBH':0,'KDQ':0,'KBU':0,'007':0,'KEI':0,'KAC':0,'KAJ':0,'KBZ':0,'KCS':0,'KBV':0,'KCF':0,'KDC':0,'KAY':0,'KBF':0,'KAS':0,'KAD':0,'KDG':0,'KAW':0,'KCM':0,'KCA':0,'KDF':0,'KEF':0,'KEY':0,'KAH':0,'KAF':0,'KEL':0,'KEW':0,'013':0,'KAR':0,'KAV':0,'KAO':0,'KAQ':0,'KCI':0,'KAP':0,'KCH':0,'KAI':0,'KAM':0,'KAL':0,'KEN':0,'KFT':0,'KFK':0,'KGX':0,'KFJ':0,'KGV':0,'KFS':0,'KFF':0,'KFH':0,'KGW':0,'KFI':0,'KAK':0,'KEG':0,'KBM':0,'KEJ':0,'KDM':0,'KEZ':0,'KBS':0,'KFM':0,'KHO':0,'KHN':0,'KHM':0,'KBY':0,'KBW':0,'KDD':0,'KDP':0,'KDS':0,'KCT':0,'KDU':0,'KCE':0,'KHQ':0}

# cnl: Channel
data['cnl'] = data['channel'].replace(channelscalemapper) 




#-----------------------------------------------------------

# 7: 'Is_Foreigner' Column:
# Most of the clients are not foreign. As seen 95.45% of the clients are Spanish only 4.54 % is Foreign and there is NaN Data
# There are very small amount of NaN values which is around 0.5%. This will also melt in binarization.
# N: Nop  Nope means No in Spanish 
# S: S or si means Yes in Spanis

#isc: is Spanish Client
data['isc'] = (data['Is_Foreigner'] == 'N').astype(int)


#-----------------------------------------------------------

# 8:'Cust_Relship_Type_Mth_Beg' Column
# The value distribution is very clear. So basically P and R values are not important. There are 10 P and 2 R.
# Value distribution is pretty clear and it is better to prefer binarization. I will be 1 A will be 0 
# The Value Distribution 
# I    0.50.6315
# A    0.49.3554
# P    0.00109%
# R    0.00022%

# iit: is I Type
data['iit'] = (data['Cust_Relship_Type_Mth_Beg'] == 'I').astype(int)


#-----------------------------------------------------------

# 9: 'Age' Column:
# There are 0.50% NaN Values in the Age and also as I see Age data is not really updated.
# The youngest client is 2 and the oldest client is around 112 years old.

# First I preferd to fill NaN values via categorized means
data['Age'] = data.groupby(['Gross_HHLD_Income', 'Cust_Province_Name', 'Cust_Sex'])['Age'].transform(lambda x: x.fillna(x.mean()))


# It can be good to create another feature and categorize Age vaues
# Categorization criterias

# Created Age Groups 

# Age Group Metodologies:
# Age Interval                   Age Grop
# 0 - 14                          Child
# 14 - 18                         Teen
# 18 - 15                         Young Adult
# 70 - 120                        Adult

# cad: Categorized Age Data
data['cad'] = pd.cut(x=data['Age'], bins = [0, 14, 18, 25, 70, 120], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

#-----------------------------------------------------------

# 10: 'Cust_Sex' Column

# H: Hombre means man transformed to 1 rest is 0 

#bgd: Binarized Gender Data
data['bgd'] = (data['Cust_Sex'] == 'H').astype(int)


#-----------------------------------------------------------

# 11: 'e-Acct' Column:
# It is one of the Target/Label Columns also does not contain any NaN value
# No any further process performed
# The data is binary data which contains only 0 and 1 

#-----------------------------------------------------------

# 12: 'Taxes' Column

# Also Taxes is one of the Target/Label Columns also does not contain any NaN value
# No any further process performed
# The data is binary data which contains only 0 and 1 


#-----------------------------------------------------------

# 13: 'Credit_Card' Column
# It is one of the Target/Label Columns also does not contain any NaN value
# No any further process performed
# The data is binary data which contains only 0 and 1 


#-----------------------------------------------------------

# 14: 'Securities' Column

# It is one of the Target/Label Columns also does not contain any NaN value
# No any further process performed
# The data is binary data which contains only 0 and 1 


#-----------------------------------------------------------

# 15: 'Pension2' column
# Pension2 is one of the Target/Label Columns and contains 232 NaN values.
# No Estimation or filling applied. All NaN dropped
# The data is binary data which contains only 0 and 1 

data['Pension2'].dropna(inplace=True)
#data['Pension2'] = data['Pension2'].astype(int)

#-----------------------------------------------------------



data = data.dropna()

pds = data.shape[0]

# dlr: Data Lose Rate 
dlr = (uds - data.shape[0]) / uds


print('-------------------------------------------------------')
print('-------------------------------------------------------')
print()
print()
print()
print()
print('Regarding the PreProcessing, the data transformed and totally {:.2%} of untransformed NaN data dropped'.format(dlr))
print()
print('The data is alligable only for analysis and analytics currently.')
print()
print()
print()
print('-------------------------------------------------------')
print('-------------------------------------------------------')
print()
print()
print()
```

## PostProcess

I did created a simple algorithm to label the data. 

The algorithm labels the changes of 5 given parameters comparing previous data.

The comparison made by recent change in order to see results of multiple decision changes of the clients. If client takes one product it turns from 0 to 1 next month but if it drops it turns. If client does not change the decision it stays as 0 but if client drops a product it returns to -1




```python
# CPDLA: Customer Preferences Differenciation Labeling Algorithm 

def CPDLA(n, a=2):
    # n dimension 
    n = n
    
    # a: appearance
    a = a
    
    #aid: Appearance in DataFrame 
    
    n = n.sort_values(['Cust_ID', 'Status_Dt'])
    
    n['cnl'] = pd.to_numeric(n['cnl'])
    n['epd'] = pd.to_numeric(n['epd'])
    
    
    
    n['aid'] = n.groupby('Cust_ID')['Cust_ID'].transform('count')
    
    #aar: Appearance Ascending Rank 
    n['aar'] = n.groupby('Cust_ID')['Status_Dt'].rank(method='dense', ascending=True)
    
    #adr: Appearance Descending Rank
    n['adr'] = n.groupby('Cust_ID')['Status_Dt'].rank(method='dense', ascending=False)
    
    # Choosing minimum 2 time appeared clien in order to watch the changes
    n = n[n['aid'] > a]
    
    # Created Lag values in order to compare changes in clients' preferences
    #Date lag is not functional I only created for the reviewer to help following the code
    n['lagdate'] = n['Status_Dt'].shift(1)
    n['lagdate'] = n['lagdate'].fillna(method='bfill')
    
    #lag e-Acct 
    n['lage-Acct'] = n['e-Acct'].shift(1)
    n['lage-Acct'] = n['lage-Acct'].fillna(method='bfill')
    
    #lag Taxes 
    n['lagTaxes'] = n['Taxes'].shift(1)
    n['lagTaxes'] = n['lagTaxes'].fillna(method='bfill')
    
    #lag Credit Card
    n['lagCredit_Card'] = n['Credit_Card'].shift(1)
    n['lagCredit_Card'] = n['lagCredit_Card'].fillna(method='bfill')
    
    #lag Securities
    n['lagSecurities'] = n['Securities'].shift(1)
    n['lagSecurities'] = n['lagSecurities'].fillna(method='bfill')
    
    #lag Pension2 
    n['lagPension2'] = n['Pension2'].shift(1)
    n['lagPension2'] = n['lagPension2'].fillna(method='bfill')
  
    # cieAL: Change in e-Acct Label
    n['cieAL'] = n['e-Acct'] - n['lage-Acct'] 
    
    # ciTxL: Change in Taxes Label
    n['ciTxL'] = n['Taxes'] - n['lagTaxes'] 

    # ciCCL: Change in Credit Cards Label
    n['ciCCL'] = n['Credit_Card'] - n['lagCredit_Card'] 
    
    # ciScL: Change in Securities Label
    n['ciScL'] = n['Securities'] - n['lagSecurities'] 
    
    # ciP2L: Change in Pension2 Label
    n['ciP2L'] = n['Pension2'] - n['lagPension2'] 
    
    # Choosing the last apperance of the client
    #n = n[n['adr'] == 1]
    
    n = n[['Cust_ID', 'Gross_HHLD_Income', 'Age', 'epd', 'cnl', 'isc', 'iit', 'bgd', 'aid', 'cieAL', 'ciTxL', 'ciCCL', 'ciScL', 'ciP2L']]
    
    
    
    n = n.set_index('Cust_ID')
    
    return n
```

## Modeling


I came across in your tamplate dataset with following notifiction:

***Beware! a random train,test split will not work here, as we will leak data to future occurances.***

***We will take samples where the next month in the last month in the db as our testing data, and the rest as train.***

This modeling concept currpts the prediction. **Why?**  

Because there are many different months and many different age, gender and income groups. I can apply your methodology and use ** - hight customized and optimized - DecisionTreeClassifier** and can have very high accuracy; but in reality it will just create serious overfitting and model curruption in real life.

So the task and dominanted/required model concept is not consistent. 


If you want to see client behavior and preferences change the most stable ways; 

- content based recommender

Client generally writes preferences and via basic NLTK and lemmatization client explanation are vectorized and than it is being used for reccomender sytem.

- CF: Collobrative Filtering 

The dta is not alligable for this. 


So basically I can generally aggregate the data and predict by time chages and general tendencies, or clients behaviors. 


** So using last month as test and previous months as training will not provide any added value ** There is only two different option.

1. Taking only last month's behavior (preferences of the clients) 

2. Taking all months and using general training - test split method. 

I'm not sure keeping the same client in the dataset more than 1 (one) time is a good idea? Because if there would be external data such as like inflation, interest rate, bank commission, foreign exhange currency, etc. It could be logical because in the end the same client could behavior different in different conditions.

There is no external value in the dataset. So I believe with exist dataset the best think we can do is predict clients's last choosie. 

There is also two main approach here; *comparing the first month and last month only* and the second approach is *comparing all the months with previous months* Some clients appears 10 times some only 1 time (which I dropped) and some 2, 3, 4 times. 

So generally if the analysis is static; which means quarterly and/or anual generally only the first month and last month comparisan is used but in dymenic analysis all movements are being used. So I use second approach here.
